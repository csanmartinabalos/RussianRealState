```python
import csv
import requests
from bs4 import BeautifulSoup

# Base URL
base_url = "https://www.geonames.org/search.html?q=region&country=RU&page="

# Número total de páginas (ajústalo según sea necesario)
num_pages = 1558

# Crear un archivo CSV para escribir los resultados
with open('ciudades_lat_long.csv', 'w', newline='', encoding='utf-8') as csvfile:
    # Crear un objeto CSV
    csv_writer = csv.writer(csvfile)

    # Escribir la cabecera del archivo CSV
    csv_writer.writerow(['Ciudad', 'Latitud', 'Longitud'])

    # Iterar sobre las páginas
    for page_num in range(1, num_pages + 1):
        # Construir la URL de la página actual
        url = base_url + str(page_num)

        # Realizar la solicitud HTTP
        response = requests.get(url)

        # Verificar si la solicitud fue exitosa (código de estado 200)
        if response.status_code == 200:
            # Analizar el contenido HTML de la página
            soup = BeautifulSoup(response.text, 'html.parser')

            # Encontrar la tabla que contiene la información de las ciudades
            table = soup.find('table', {'class': 'restable'})

            # Iterar sobre las filas de la tabla
            for row in table.find_all('tr')[1:]:
                # Obtener los datos de la fila
                columns = row.find_all('td')

                # Verificar si hay suficientes columnas en la fila
                if len(columns) >= 6:
                    # Extraer el nombre de la ciudad, latitud y longitud
                    ciudad = columns[1].text.strip()
                    latitud = columns[4].text.strip()
                    longitud = columns[5].text.strip()

                    # Escribir los datos en el archivo CSV
                    csv_writer.writerow([ciudad, latitud, longitud])
                else:
                    print(f"Fila sin suficientes columnas en la página {page_num}. Se omite.")

            print(f"Extracción exitosa de la página {page_num}.")

        else:
            print(f"Error al obtener la página {page_num}. Código de estado:", response.status_code)
```

```python
import pandas as pd

# URL de la página de Wikipedia
url = "https://en.wikipedia.org/wiki/List_of_cities_and_towns_in_Russia"

# Leer todas las tablas HTML de la página
tables = pd.read_html(url, header=0)

# Seleccionar la tabla que contiene la información de las ciudades
selected_table = tables[0]

# Seleccionar las columnas "Name" y "City"
selected_columns = selected_table[["Federal subject", "City"]]

# Guardar los datos en un archivo CSV
selected_columns.to_csv('ciudades_rusia.csv', index=False, encoding='utf-8')

print("Extracción y guardado exitosos.")

```
